```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
data = pd.read_csv('50_Startups.csv')
```

```python
data.head()
```

## Bu veri seti, 50 farklı startup şirketinin çeşitli harcamalarını ve kârlılıklarını içermektedir.  

R&D Spend (Ar-Ge Harcaması): Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  
Administration (Yönetim Harcaması): Şirketin yönetim giderleri için harcadığı tutar.  
Marketing Spend (Pazarlama Harcaması): Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar. 
State (Eyalet): Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  
Profit (Kâr): Şirketin elde ettiği toplam kâr.  
Bu veri seti, startup'ların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz   etmek için kullanılabilir. Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini   incelemek için uygun bir veri setidir.  

## 1.GÖREV : R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
# Kodu buraya yazınız 

plt.figure(figsize=(10, 6))

# Scatter Plot Dağılım Grafiği Çizimi

plt.scatter(data['R&D Spend'], data['Profit'], color='blue', alpha=0.7)

# Başlık ve Etiketler
plt.title('Ar-Ge Harcaması ve Kâr İlişkisi', fontsize=15)
plt.xlabel('Ar-Ge Harcaması (R&D Spend)', fontsize=12)
plt.ylabel('Kâr (Profit)', fontsize=12)

plt.grid(True, linestyle='--', alpha=0.5)

# Grafiği göster
plt.show()




```

## 2.GÖREV: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot): Yönetim harcamaları ile kâr arasındaki ilişkiyi gösteren bir dağılım grafiği.

```python
# Kodu buraya yazınız 

plt.figure(figsize=(10, 6))


plt.scatter(data['Administration'], data['Profit'], color='red', alpha=0.6)

# 3. Başlık ve Etiketler
plt.title('Yönetim Harcamaları ve Kâr İlişkisi', fontsize=15)
plt.xlabel('Yönetim Harcamaları (Administration)', fontsize=12)
plt.ylabel('Kâr (Profit)', fontsize=12)

# Izgara çizgileri
plt.grid(True, linestyle='--', alpha=0.5)

# 4. Göster
plt.show()




```

## 3. GÖREV: Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startup'ların ortalama kârlarını karşılaştıran bir çubuk grafik.

```python
# Kodu buraya yazınız 
ortalama_karlar = data.groupby('State')['Profit'].mean()


print("Eyaletlere Göre Ortalamalar:\n", ortalama_karlar)


plt.figure(figsize=(10, 6))

renkler = ['red', 'green', 'blue'] 
plt.bar(ortalama_karlar.index, ortalama_karlar.values, color=renkler)

# Başlık ve Etiketler
plt.title('Eyaletlere Göre Ortalama Startup Kârları', fontsize=15)
plt.xlabel('Eyalet', fontsize=12)
plt.ylabel('Ortalama Kâr', fontsize=12)


plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()




```

## 4. GÖREV: Harcama Türlerinin Karşılaştırması (Boxplot): R&D, yönetim ve pazarlama harcamalarının dağılımını karşılaştıran bir kutu grafiği.

```python
# Kodu buraya yazınız 

plt.figure(figsize=(10, 6))

veriler = [data['R&D Spend'], data['Administration'], data['Marketing Spend']]
isimler = ['Ar-Ge', 'Yönetim', 'Pazarlama']

kutular = plt.boxplot(veriler, labels=isimler, patch_artist=True)

renkler = ['lightblue', 'lightgreen', 'pink']
for kutu, renk in zip(kutular['boxes'], renkler):
    kutu.set_facecolor(renk)


plt.title('Harcama Kalemlerinin Karşılaştırması', fontsize=15)
plt.ylabel('Harcama Miktarı ($)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)

# 5. Göster
plt.show()




```
